Digitar no prompt de comando:

__ver versão instalada:__
>python

__criar um ambiente com o nome python2 com uma versão diferente do python__
>conda create --name python2 python=2.7

__ativar o ambiente para ser usado__
>activate python2

__desativar o ambiente. faz voltar para a versão (base)__
>deactivate

__ver todos os ambientes já instalados__
>conda info --envs
