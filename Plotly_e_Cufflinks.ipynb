# Plotly e Cufflinks
São bibliotecas iterativas, que rederizam gráficos que podemos interagir com eles, dá mais flexibilidade na análise.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline   
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
import cufflinks as cf
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(100, 4), columns="A B C D".split())
df1
```

```python
df2 = pd.DataFrame({"categoria": ["A", "B", "C"], "Valores": [12, 45, 12]})
df2
```

```python
# plot com iplot do plotly
df1.iplot(kind="scatter", x="A", y="B", mode="markers")
```

```python
df1[["A", "B"]].iplot(kind="spread")
```

```python
df1["A"].iplot(kind="hist")
```
