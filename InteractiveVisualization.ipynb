```python
import seaborn as sns
import matplotlib.pyplot as plt
mpg = sns.load_dataset('mpg')

def f(x):
    img = sns.countplot(mpg[mpg['mpg']<x]['origin']) 
    plt.title('Origin Countplot mpg = {}'.format(x))
    return img
```
