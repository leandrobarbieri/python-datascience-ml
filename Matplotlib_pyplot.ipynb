# Matplotlib.pyplot
Biblioteca padrão para visualização de dados, usada pelas outras como base para os gráficos.

```python
import numpy as np
import matplotlib.pyplot as plt

# tamanho padrão da fig caso não seja especificado
plt.rc('figure', figsize=(5, 3))

# visualizar os graficos criados na propria linha de codigo usada para definição
# não precisar escrever plt.show()
%matplotlib inline
```

# Sem criar figures

```python
# dados de testes para usar nos eixos dos graficos
x = np.linspace(0, 5, 11)
y = x * x
```

```python
# Forma mais direta para criar um grafico
plt.plot(x, y, color="red")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Titulo")
```

```python
# alterando o tipo para scatter/bar/step/fill_between
plt.scatter(x, y, color="red")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Titulo")
```

## Subplots sem criar figures

```python
# subplots: 1
plt.subplot(1, 2, 1)
plt.plot(x, y, color="red")

# subplots: 2
plt.subplot(1, 2, 2)
plt.plot(y, x, color="blue")
```

# Utilizando figure

## adicionando axes

```python
# fig caixa vazia para ter controle total sobre a figura e subplots
fig = plt.figure()

# posições em percentual das extremidade
axes1 = fig.add_axes([0.1, 0.1, 0.5, 1])
axes2 = fig.add_axes([0.7, 0.1, 0.5, 1])

# setando propriedades do axes1
axes1.set_xlabel("Eixo-X1")
axes1.set_ylabel("Eixo-Y1")
axes1.set_title("Titulo1")
axes1.plot(x, y)

# setando propriedades do axes2
axes2.set_xlabel("Eixo-X2")
axes2.set_ylabel("Eixo-Y2")
axes2.set_title("Titulo2")
axes2.plot(x, y)
```

# criando subplots

```python
# criando varios subplots
fig, ax = plt.subplots()

# mais de uma linha em um mesmo subplot
ax.plot(x, x**3, "b--")
ax.plot(x, x**4, "g--")

ax.set_xlabel("EixoX")
ax.set_ylabel("EixoY")
ax.set_title("Titulo")
```

```python
# retorna uma matriz de objetos subplots
fig, ax = plt.subplots(nrows=2, ncols=3)

# acessando a primeira
ax[0][0].plot(x, y*y, "r--") 
ax[0][0].set_xlabel("Label1")

ax[0][1].plot(x, y*y, "r--") 
ax[0][1].set_xlabel("Label1")

ax[0][2].plot(x, y*y, "r--") 
ax[0][2].set_xlabel("Label1")

# acessando a segunda
ax[1][0].plot(x, y*2, "b--") 
ax[1][0].set_xlabel("Label2")

ax[1][1].plot(x, y*2, "b--") 
ax[1][1].set_xlabel("Label2")

ax[1][2].plot(x, y*2, "b--") 
ax[1][2].set_xlabel("Label2")

# Ajusta os subplots para evitar sobreposições e melhorar a apresentaçao
plt.tight_layout()
```

```python
# Ajustando o tamanho 
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,2))
axes.plot(x, x, "r--", label="Red (x - x)")
axes.plot(y, x, "g--", label="Green (x - y)")
axes.set_title("Titulo1")

# loc=0 o matplotlib escolhe o melhor lugar
axes.legend(loc=0)
```

## Salvando o grafico como imagem

```python
# salva a imagem como o nome e o formato escolhido no diretorio do jupyter
fig.savefig("imagem.jpg")
```

```python
# cria uma matriz de subplots com 2 linhas e 2 colunas
# ajusta o tamanho da figura para 16,3
# sharey=True mantém o mesmo rotulo de eixo y para todos os graficos na mesma linha da matriz de subplots
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(16,3), sharey=True)


axes[0].plot(x, y, color="red",  linewidth=0.5, linestyle=":", marker="+", markersize=5, markerfacecolor="yellow")
axes[0].set_xlim([0, 10])
axes[0].set_ylim([0, 30])
axes[0].set_title("Titulo1")
axes[0].set_ylabel("Eixo compartilhado")

# cores com nome e formato hex
axes[1].set_title("Titulo2")
axes[1].plot(x, y, color="#8B8B8B", linewidth=10, alpha=0.1, linestyle="--", marker="o", markersize=16, markerfacecolor="red")
axes[1].grid(color='#8B8B8B', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

## Dois eixos

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")

# itera sob os rotuos do eixo y
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
# cria o eixo secundario, compartilhando o mesmo eixo x 
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")

# itera sob os rotuos do eixo y
for label in ax2.get_yticklabels():
    label.set_color("red")
```

# Outros tipos graficos

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

## Anotações no grafico

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, x, x**3)

ax.text(5, 20, r"$y=x^2$", fontsize=20, color="blue")
ax.text(5, 100, r"$y=x^3$", fontsize=20, color="orange");
```
