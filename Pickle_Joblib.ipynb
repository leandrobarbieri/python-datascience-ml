# Pickle
Salvando o modelo com pickle

```python
# Salvar um modelo usando pickle
# https://machinelearningmastery.com/save-load-machine-learning-models-python-scikit-learn/
# https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv

import pandas
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(url, names=names)
dataframe
```

```python
array = dataframe.values

# todas as primeiras 7 colunas (features)
X = array[:,0:8]

# classe target
Y = array[:,8]

# separação do treino e teste
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.33, random_state=1)

# Fit
model = LogisticRegression()
model.fit(X_train, Y_train)

# save the model to disk
filename = 'pickle_model.sav'

# Descarrega o modelo em um arquivo binário
pickle.dump(model, open(filename, 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, Y_test)
print(result)
```

# Joblib
Salvando o modelo com joblib

```python
import joblib
filename = 'joblib_model.sav'

# salva o modelo como arquivo
joblib.dump(model, filename)
```

```python
# load the model from disk
loaded_model = joblib.load(filename)
result = loaded_model.score(X_test, Y_test)
print(result)
```
